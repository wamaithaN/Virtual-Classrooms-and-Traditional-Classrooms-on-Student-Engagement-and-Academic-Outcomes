```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("C:/Users/PC/Downloads/Virtual vs Traditional Student Engagement (1).xlsx")
data
print(data.columns.tolist())
```

```python
# Descriptive statistics
print(data[['Engagement Levels (Virtual)', 'Engagement Levels (Traditional)']].describe())

# Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['Engagement Levels (Virtual)', 'Engagement Levels (Traditional)']])
plt.title('Engagement Levels: Virtual vs Traditional')
plt.show()
```

```python
# Correlation heatmap
corr = data[['Engagement Levels (Virtual)', 'Engagement Levels (Traditional)', 'Academic Outcomes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()


# Scatterplot
sns.scatterplot(data=data, x='Engagement Levels (Virtual)', y='Academic Outcomes', hue='Gender ')
plt.title('Academic Outcomes vs Virtual Engagement')
plt.show()
```

```python
# Satisfaction levels distribution
sns.histplot(data=data, x='Student Satisfaction Level', bins=8, kde=True, hue='Gender ')
plt.title('Student Satisfaction Levels')
plt.show()

# Cognitive load vs satisfaction
sns.scatterplot(data=data, x='Cognitive Load (N-Back)', y='Student Satisfaction Level', hue='Gender ')
plt.title('Cognitive Load vs Student Satisfaction')
plt.show()
```

```python
# Interaction and engagement relationship
sns.barplot(data=data, x='Teacher-Student Interaction Frequency', y='Academic Outcomes', hue='Gender ')
plt.title('Interaction Frequency vs Academic Outcomes')
plt.show()

# Pairplot for related variables
sns.pairplot(data[['Teacher-Student Interaction Frequency', 'Academic Outcomes', 'Engagement Levels (Virtual)']])
plt.suptitle('Pairplot Analysis', y=1.02)
plt.show()
```

```python
# Engagement levels by gender
sns.boxplot(data=data, x='Gender ', y='Engagement Levels (Virtual)')
plt.title('Engagement Levels (Virtual) by Gender')
plt.show()

sns.boxplot(data=data, x='Gender ', y='Engagement Levels (Traditional)')
plt.title('Engagement Levels (Traditional) by Gender')
plt.show()
```
